# Bienvenido al curso de Python

El objetivo de este curso es **iniciarte desde cero en la programación en Python y aprender distintas aplicaciones de este lenguaje en la ingeniería**.
Nuestra herramienta fundamental de trabajo es el **Cuaderno (Notebook) de Jupyter** (este documento ha sido generado a partir de un notebook).

En esta sesión inicial, veremos los pasos a seguir para **descargar el material e instalar Python**. 

## Descarga de los cuadernos 

Los cuadernos AERO y datos requeridos (archivos de audio, imagen y video) se encuentran contenidos en el archivo comprimido `clases.zip` accesible desde 
<div class="alert alert-block alert-warning">

Descomprima el archivo y almacenelo en su PC local.

## Instalacion de Miniconda

[Conda](https://conda.io/docs/) es un sistema de gestión de paquetes de código abierto el cual instala, ejecuta y actualiza rápidamente los paquetes y sus dependencias. El mismo corre en Windows, macOS and Linux. El paquete Conda se incluye en **Anaconda** y en su version minima **Miniconda**. Este ultimo permite instalar Python, Jupyter y todas las librerias requeridas por el curso.

Descargue e instale la version de Miniconda acorde a su sistema operativo desde el siguiente sitio web:
<div class="alert alert-block alert-warning">
<a href="https://docs.conda.io/en/latest/miniconda.html">sitio Miniconda</a>
</div> 

## Creacion del entorno AERO

Proveemos un archivo de entorno el cual presenta una lista con todos los paquetes requeridos para interactuar con los cuadernos AERO. Abra una ventana de comandos en Windows o terminal en Linux/macOS. Desde el directorio que contiene el archivo de entorno `environment.yml` ejecute lo siguiente: 
<div class="alert alert-block alert-warning">
<tt>conda env create -f environment.yml</tt>
</div> 

## Lanzamiento del servidor Jupyter 

A continuacion, activamos el entorno (*ambiente*) AERO con el siguiente comando:

<div class="alert alert-block alert-warning">
<tt>conda activate AERO</tt>
</div> 

Finalmente, desde el directorio que contiene los cuadernos AERO
ejecute lo siguiente para abrir una sesión de Jupyter en su navegador web:

<div class="alert alert-block alert-warning">
<tt>jupyter notebook</tt>
</div> 

Este comando permitira mostrar la estructura de los cuadernos. Asi, sera posible abrir cualquier cuaderno seleccionando un archivo con la extension `.ipynb`.
